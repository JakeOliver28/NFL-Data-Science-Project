# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
import numpy as np
```

```python
team_df = pd.read_csv('/data/JakeOliver28/teams.csv')
team_df.head()
```

```python
plt.hist(team_df['WinPercentage'])
plt.xlabel('Winning Percentage')
plt.ylabel('Count')
plt.title('Winning Percentage')
plt.show()
```

### Passing Yards

```python
plt.scatter('PassYards', 'WinPercentage', data=team_df)
plt.xlabel('PassYards')
plt.ylabel('WinPercentage')
plt.show()
print("Mean: {}".format(np.mean(team_df['PassYards'])))
print("Standard Deviation: {}".format(np.std(team_df['PassYards'])))
print("Min: {}".format(np.min(team_df['PassYards'])))
print("Max: {}".format(np.max(team_df['PassYards'])))
```

### Rush Yards

```python
plt.scatter('RushYards', 'WinPercentage', data=team_df)
plt.xlabel('RushYards')
plt.ylabel('WinPercentage')
plt.show()
print("Mean: {}".format(np.mean(team_df['RushYards'])))
print("Standard Deviation: {}".format(np.std(team_df['RushYards'])))
print("Min: {}".format(np.min(team_df['RushYards'])))
print("Max: {}".format(np.max(team_df['RushYards'])))
```
